```python
import pickle

no_c_file = '2023-11-05 genre no c.pkl'
c_file = '2023-11-07 genre c.pkl'

# Open the file in binary read mode
with open(no_c_file, 'rb') as file:
    # Load the contents from the file
    no_c = pickle.load(file)

# Open the file in binary read mode
with open(c_file, 'rb') as file:
    # Load the contents from the file
    c = pickle.load(file)
```

```python
full = {**no_c, **c}
# Open the file in binary write mode
with open('genre.pkl', 'wb') as file:
    # Dump the `your_data` object to the file
    pickle.dump(full, file)
```

```python
def print_results(raw_text, output):
    for start, end, entity in output:
        print(start, end, entity)
        print(raw_text[start:end])
        print(raw_text[start-10:start] + "*" + raw_text[start:end] + "*" + raw_text[end:end+10])
        print("")
```

```python
for raw_text, output in c.items():
    print("---------")
    print_results(raw_text, output)
```
